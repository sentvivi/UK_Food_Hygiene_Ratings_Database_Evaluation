5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
# Convert latitude and longitude strings to Decimal
update_query = [
    {
        "$set": {
            "geocode.longitude": {
                "$convert": {
                    "input": "$geocode.longitude",
                    "to": "decimal"
                }
            },
            "geocode.latitude": {
                "$convert": {
                    "input": "$geocode.latitude",
                    "to": "decimal"
                }
            }
        }
    }
]

# Update documents to change data types
establishments.update_many({}, update_query)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
establishments.update_many({"RatingValue": {"$type": "string"}}, [{"$set": {"RatingValue": {"$toInt": "$RatingValue"}}}])
```

```python
# Check that the coordinates and rating value are now numbers
document = establishments.find_one()
# Display data types of coordinates and RatingValue
if document:
    longitude_type = type(document.get('geocode', {}).get('longitude'))
    latitude_type = type(document.get('geocode', {}).get('latitude'))
    rating_value_type = type(document.get('RatingValue'))

    print(f"Longitude data type: {longitude_type}")
    print(f"Latitude data type: {latitude_type}")
    print(f"RatingValue data type: {rating_value_type}")
else:
    print("No documents found.")
```
